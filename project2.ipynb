## Part 2: Explore one Wikipedia page with the beautifulsoup package

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#save and print the text content of a page with all tags removed

url = "https://en.wikipedia.org/wiki/Chicago"
source=requests.get(url).text
soup = BeautifulSoup(source)
print(soup.get_text())
```

```python
#download an image with beautifulsoup and save it in this repository

img_tags=soup.find_all('img')
img_tags[1]['src']
```

```python
#download image from link
img_url=img_tags[0]['src']
img_url = 'https://wikipedia.org'+img_url
img=requests.get(img_url)
```

```python
#save image to file
with open('wiki_logo.jpg','wb') as f:
    f.write(img.content)
```

```python
#find all the links in a page with beautifulsoup
#print the first 100 characters of ten of these links
```

```python
#make dataframe with all elements with link and class container__link
links=soup.find_all('a')    
link_url=[]
for i in links:

    #issues with some links not having href, we just use try except to skip it
    try:
        link_url.append(i['href'])
    except:
        continue
print(link_url)
```

```python
#we can see that some links are not complete, we can add the base url to them
link_url=[i if 'http' in i else 'https://wikipedia.org'+i for i in link_url][:10]
```

```python
link_url
```

```python
#cleaning function to remove all whitespace and newlines
def clean(text):

    #convert the text to lowercase
    text = text.lower()
    #remove punctuation
    text = ''.join([char for char in text if char.isalpha() or char.isspace()])
    #remove newline and space characters
    text = ' '.join(text.split())
    return text
```

```python
#we can get the text of all the links
link_text=[]
title=[]
count=0
for i in link_url:
    source=requests.get(i).text
    soup=bs4.BeautifulSoup(source)
    link_text.append(clean(soup.text))
    title.append(clean(soup.title.text))

    #run first ten to save time
    count+=1
    if count==10:
        break
df=pd.DataFrame({'title':title,'text':link_text})
```

```python
#using the links in the "link" column, download the first 1000 characters of each script
#use requests and bs4, remember to remove all html tags

def get_scripts(url):
    try:
        response = requests.get(url)
        soup = bs4.BeautifulSoup(response.text)
        text = soup.get_text()
        return text[:1000]
    except:
        return None
    
scripts['script'] = scripts['link'].apply(get_scripts)
```
